```python
##Question 1
## Write a function for translating DNA to protein

def translate(sequence): 
    table = { 
        'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M', 
        'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T', 
        'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K', 
        'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',                  
        'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L', 
        'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P', 
        'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q', 
        'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R', 
        'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V', 
        'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A', 
        'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E', 
        'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G', 
        'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S', 
        'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L', 
        'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_', 
        'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W', 
    } 
    protein = "" 

    # Ensure the sequence length is a multiple of 3
    if len(sequence) % 3 != 0:
        print("Error: The sequence is not a multiple of three. Ignoring the extra bases.")
        sequence = sequence[:len(sequence) - (len(sequence) % 3)]  # Trim extra bases

    for i in range(0, len(sequence), 3): 
        codon = sequence[i:i + 3]
        if codon in table:
            protein += table[codon] 
        else:
            raise ValueError(f"Invalid codon found: {codon}")

    return protein 

# Test Cases
print(translate("ACTCGA"))  
print(translate("ACTCG"))   # Trims the last 2 bases and prints a warning
```

```python
###Question 2: Write a function that simulates and generates a logistic population growth curve. 
##Your function should include 2 extra parameters that randomize the length of the lag phase and the exponential phase 
## Most living populations follow a logistic population growth. 
##Therefore, your growth curve can be: Population Size vs Time, Cell density vs Time, OD vs Time, CFU vs Time, etc
##Using your function, generate a dataframe with 100 different growth curves

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Logistic Growth Function
def logistic_growth(t, K, r, t0):
    return K / (1 + np.exp(-r * (t - t0)))

# Function to generate a logistic growth curve with randomized lag and exponential phases
def generate_growth_curve(time_points=100, K=1.0, r=0.5):
    t = np.linspace(0, 50, time_points)  # Time from 0 to 50 units
    lag_phase = np.random.uniform(5, 15)  # Random lag phase length
    exp_phase = np.random.uniform(5, 15)  # Random exponential phase length
    t0 = lag_phase + exp_phase / 2  # Inflection point (midpoint of growth)
    
    population = logistic_growth(t, K, r, t0)
    return t, population

# Generate 100 different growth curves
data = []
num_curves = 100

for i in range(num_curves):
    t, pop = generate_growth_curve()
    df = pd.DataFrame({'Time': t, f'Population_{i}': pop})
    data.append(df.set_index('Time'))

# Combine into a single DataFrame
growth_df = pd.concat(data, axis=1)

```

```python
### Question 3.  Finally, write a function for calculating the hamming distance between your Slack username and twitter/X handle (synthesize if you donâ€™t have one). 
##Feel free to pad it with extra words if they are not of the same length.


def hamming_distance(string1, string2): 
    # Start with a distance of zero, and count up
    distance = 0
    # Loop over the indices of the string
    L = len(string1)
    for i in range(L):
        # Add 1 to the distance if these two characters are not equal
        if string1[i] != string2[i]:
            distance += 1
    # Return the final count of differences
    return distance

slack_username = "Abdulkabir11111"
twitter_handle = "Omeiza_PlantDoc"

hamming_distance(slack_username, twitter_handle)
```

```python
##Write a function for determining the time to reach 80% of the maximum growth;
##usually the carrying capacity


def time_to_reach_80_percent_K(time, population, K=1.0):
    threshold = 0.8 * K
    return time[np.argmax(population >= threshold)]

# Example usage:
example_t, example_pop = generate_growth_curve()
time_80 = time_to_reach_80_percent_K(example_t, example_pop)

# Plot an example
plt.plot(example_t, example_pop, label='Growth Curve')
plt.axhline(0.8, linestyle='--', color='r', label='80% of K')
plt.axvline(time_80, linestyle='--', color='g', label='Time to 80% K')
plt.xlabel("Time")
plt.ylabel("Population Size")
plt.legend()
plt.show()

growth_df.head()
```
